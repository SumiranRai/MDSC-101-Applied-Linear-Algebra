```python
def matrix_addition(A, B):
    # Check dimensions
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        raise ValueError("Matrices must have the same dimensions for addition.")

    # Initialize result matrix with zeros
    rows = len(A)
    cols = len(A[0])
    result = [[0]*cols for _ in range(rows)]

    # Perform addition element-wise
    for i in range(rows):
        for j in range(cols):
            result[i][j] = A[i][j] + B[i][j]

    return result

# Function to input a matrix of given dimensions from the user
def input_matrix(prompt, rows, cols):
    matrix = []
    print(prompt)
    for i in range(rows):
        while True:
            row = input(f"Enter row {i+1} of {rows} (space-separated values): ").strip()
            if row:
                break
        matrix.append([int(x) for x in row.split()])
    return matrix

# Input dimensions for matrices from the user
rows = int(input("Enter the number of rows for matrices A and B: "))
cols = int(input("Enter the number of columns for matrices A and B: "))

# Input matrices A and B from the user
print("Enter matrix A:")
A = input_matrix("Enter row (space-separated values), press enter for next row:", rows, cols)

print("Enter matrix B:")
B = input_matrix("Enter row (space-separated values), press enter for next row:", rows, cols)

# Perform matrix addition
try:
    result = matrix_addition(A, B)
    print("\nResult of matrix addition:")
    for row in result:
        print(row)
except ValueError as e:
    print("Error:", e)
```

```python
def gaussian_elimination(matrix):
    rows = len(matrix)
    cols = len(matrix[0])

    for i in range(min(rows, cols)):
        # Make the diagonal element 1 and eliminate the column below it
        if matrix[i][i] == 0:
            for k in range(i+1, rows):
                if matrix[k][i] != 0:
                    matrix[i], matrix[k] = matrix[k], matrix[i]
                    break
            else:
                continue

        pivot = matrix[i][i]
        for j in range(i, cols):
            matrix[i][j] /= pivot

        # Eliminate the elements below the pivot
        for k in range(i+1, rows):
            factor = matrix[k][i]
            for j in range(i, cols):
                matrix[k][j] -= factor * matrix[i][j]

    return matrix

# Function to input a matrix from the user
def input_matrix(prompt, rows, cols):
    matrix = []
    print(prompt)
    for i in range(rows):
        while True:
            row = input(f"Enter row {i+1} of {rows} (space-separated values): ").strip()
            if row:
                row_values = [float(x) for x in row.split()]
                if len(row_values) != cols:
                    print(f"Error: Expected {cols} values, got {len(row_values)}. Please enter the row again.")
                else:
                    matrix.append(row_values)
                    break
    return matrix

# Input dimensions for the matrix from the user
rows = int(input("Enter the number of rows for the matrix: "))
cols = int(input("Enter the number of columns for the matrix: "))

# Input the matrix from the user
print("Enter the matrix:")
matrix = input_matrix("Enter row (space-separated values):", rows, cols)

# Perform Gaussian elimination
result = gaussian_elimination(matrix)
print("\nMatrix after Gaussian elimination:")
for row in result:
    print(row)
```
